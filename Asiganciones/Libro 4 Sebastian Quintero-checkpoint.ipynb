```python
Curva_T=pd.read_excel('./Datos/Datos-azul-timol.xlsx', header=None, usecols=[0,1,3,5,6], nrows=20)
Curva_T
```

```python
Datos=Curva_T.drop([0])
Datos
```

```python
Datos1=Datos.drop([2,4,6,8,10,12,14,16,18], axis=0)
Datos1
```

```python
Datos2=Datos1.drop([1], axis=0)
Datos2
```

```python
Datos3=Datos2.set_axis(['Muestra','Absorbancia 1', 'Absorbancia 2', 'Absorbancia 3', 'Concentracion'], axis=1)
Datos3
```

```python
Datos3['Promedio']= Datos3[['Absorbancia 1', 'Absorbancia 2', 'Absorbancia 3']].mean(axis=1)
Datos3
```

```python
Datos4=Datos3.drop(['Absorbancia 1', 'Absorbancia 2', 'Absorbancia 3'], axis=1)
Datos4
```

```python
import sklearn.metrics as skm
from sklearn import linear_model
x1=Datos4[['Concentracion']] 
y1=Datos4[['Promedio']]
reg = linear_model.LinearRegression()
reg.fit(x1, y1)
y1_pred = reg.predict(x1)
m=reg.coef_[0]
b=reg.intercept_[0]
mse=skm.mean_squared_error(y1, y1_pred)
r2=skm.r2_score(y1, y1_pred)
```

```python
fig, ax= plt.subplots()
ax.scatter(x1, y1, color='green', label='Datos originales', edgecolors='black')
ax.plot(x1, y1_pred, 'r', label='Ajuste')
ax.legend()
ax.set_title("curva de Calibración")
ax.set_xlabel("Concentración (ppm)")
ax.set_ylabel("Intensidad promedio (units arb.)")
ax.text(0.000045,0.4, r"$r^2 = "+str("{0:.3f}".format(r2))+r"$", fontsize=12, color="black")
plt.show()
#fig.savefig("ccal1.pdf",bbox_inches='tight')
```

```python
Datos4['Desviación']= Datos3[['Absorbancia 1','Absorbancia 2','Absorbancia 3']].std(axis=1)
Datos4
```

```python
Yerror = Datos4['Desviación']
```

```python
fig, ax= plt.subplots()
ax.scatter(x1, y1, color='green', label='Datos originales', edgecolors='black')
ax.errorbar(np.array(x1), np.array(y1_pred), yerr=10*np.array(Yerror), ecolor='red')
ax.legend()
ax.set_title("Curva de Calibración")
ax.set_xlabel("Concentración (ppm)")
ax.set_ylabel("Absorbancia")
ax.text(0.00004,0.4, r"$r^2 = "+str("{0:.3f}".format(r2))+r"$", fontsize=12, color="black")
plt.show()
```
