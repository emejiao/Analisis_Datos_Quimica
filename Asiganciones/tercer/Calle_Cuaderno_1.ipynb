```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

---
<font color=black size=6 align='center'><p style='text-align: center;'>**1. DataFrames**  
    
---

```python
lugar1=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR4.CSV')
lugar2=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR6.CSV')
lugar3=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR2.CSV')
ir4=np.loadtxt(lugar1, delimiter=',')
ir6=np.loadtxt(lugar2, delimiter=',')
ir2=np.loadtxt(lugar3, delimiter=',')
lugar4=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS4.txt')
lugar5=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS6.txt')
lugar6=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS2.txt')
ms4=np.loadtxt(lugar4)
ms6=np.loadtxt(lugar5)
ms2=np.loadtxt(lugar6)
lugar7=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/Supplementary_CCR.xlsx')
sup_CCR=pd.read_excel(lugar7)
sup_Prop=pd.read_excel(lugar7, sheet_name=["properties"])
```

```python
ir1=pd.DataFrame(ir4)
ir2=pd.DataFrame(ir6)
ir3=pd.DataFrame(ir2)
ir1.set_axis(['Número de onda', 'T'], axis=1,inplace=True)
ir2.set_axis(['Número de onda', 'T'], axis=1,inplace=True)
ir3.set_axis(['Número de onda', 'T'], axis=1,inplace=True)
ms1=pd.DataFrame(ms4)
ms2=pd.DataFrame(ms6)
ms3=pd.DataFrame(ms2)
ms1.set_axis(['m/z', 'Int'], axis=1,inplace=True)
ms2.set_axis(['m/z', 'Int'], axis=1,inplace=True)
ms3.set_axis(['m/z', 'Int'], axis=1,inplace=True)
Sup_CCR1=pd.DataFrame(sup_CCR)
Sup_Prop1=pd.DataFrame([sup_Prop])
```

---
<font color=navy size=5> **1.1. Filtros y selección**</font>  

---

```python
Sup_CR= Sup_CCR1[['Molecules\Samples', 'CR1', 'CR2', 'CR3', 'CR4', 'CR5', 'CR6', 'CR7',
       'CR8', 'CR9', 'CR10', 'CR11', 'CR12', 'CR13', 'CR14', 'CR15', 'CR16',
       'CR17', 'CR18', 'CR19', 'CR20', 'CR21']]
Sup_VR= Sup_CCR1[['Molecules\Samples', 'VR1', 'VR2', 'VR3', 'VR4',
       'VR5', 'VR6', 'VR7', 'VR8', 'VR9', 'VR10', 'VR11', 'VR12', 'VR13',
       'VR14', 'VR15', 'VR16', 'VR17', 'VR18', 'VR19']]
Sup_CR.to_excel('Sup_CR.xlsx', index=False)
Sup_VR.to_excel('Sup_VR.xlsx', index=False)
Sup_Prop1.to_excel('Sup_Prop.xlsx', index=False)
Sup_CR
```

---
<font color=navy size=5> **1.2. Gráficas (plt.subplots)**</font>  

---

### <font color='8A2BE2'>Espectro IR</font>

```python
fig, ax = plt.subplots()
ax.plot(ir1['Número de onda'], ir1['T'], 'b', label='IR4') 
ax.plot(ir2['Número de onda'], ir2['T'], 'r', label='IR6') 
ax.plot(ir3['Número de onda'], ir3['T'], 'g', label='IR2') 
ax.legend(loc='lower right', fontsize=12)
ax.set_xlim(399, 4001)
ax.set_ylim(55, 102)
ax.set_title('Espectros IR')
ax.set_xlabel('Número de onda [cm$^-$$^1$]')
ax.set_ylabel('%T')
```

---
### <font color='228B22'>Espectro de masas</font>

```python
fig, ax = plt.subplots()
ax.plot(ms1['m/z'], ms1['Int'], 'b', label='MS4') 
ax.plot(ms2['m/z'], ms2['Int'], 'r', label='MS6') 
ax.plot(ms3['m/z'], ms3['Int'], 'g', label='MS2') 
ax.legend(loc='best', fontsize=12)
ax.set_xlim(100, 1000)
ax.set_ylim(0, 72000)
ax.set_title('Espectros de masas')
ax.set_xlabel('m/z')
ax.set_ylabel('Intensidad')
```

---
<font color=navy size=5> **1.3. Combinación (pd.concat)**</font>  

---

### <font color='8A2BE2'>Datos para el espectro IR</font>

```python
ir1=pd.DataFrame(ir4)
ir2=pd.DataFrame(ir6)
ir3=pd.DataFrame(ir2)
IR1=pd.concat([ir1, ir2[[1]], ir3[[1]]], axis=1)
IR1.set_axis(['Número de onda', 'T1', 'T2', 'T3'], axis=1,inplace=True)
IR1
```

```python
IR1.to_excel('IR1.xlsx', index=False)
```

### <font color='8A2BE2'>Espectro IR</font>

```python
x=IR1[['Número de onda']]
T1=IR1[['T1']]
T2=IR1[['T2']]
T3=IR1[['T3']]

plt.rcParams["font.family"] = 'Times New Roman'
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, T1,'b', label='T1')
ax.plot(x, T2,'g', label='T2')
ax.plot(x, T3,'r', label='T3')
ax.legend(loc='center right', fontsize=16)
ax.set_xlim(399, 4001)
ax.set_ylim(55, 102)
ax.set_title('Espectros IR', fontsize=20)
ax.set_xlabel('Número de onda [cm$^-$$^1$]', fontsize=16)
ax.set_ylabel('%T', fontsize=16)
ax.tick_params(labelsize=13, labelcolor='k', width=1)
plt.grid()
```

---
### <font color='228B22'>Datos para el espectro de masas</font>

```python
ms1=pd.DataFrame(ms4)
ms2=pd.DataFrame(ms6)
ms3=pd.DataFrame(ms2)
MS1=pd.concat([ms1, ms2[[1]], ms3[[1]]], axis=1)
MS1.set_axis(['m/z', 'Int 1', 'Int 2', 'Int 3'], axis=1,inplace=True)
MS1
```

```python
MS1.to_excel('MS1.xlsx', index=False)
```

### <font color='228B22'>Espectro de masas</font>

```python
#%pwd
x=MS1[['m/z']]
M1=MS1[['Int 1']]
M2=MS1[['Int 2']]
M3=MS1[['Int 3']]

plt.rcParams["font.family"] = 'Times New Roman'
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, M1,'b', label='Int 1')
ax.plot(x, M2,'g', label='Int 2')
ax.plot(x, M3,'r', label='Int 3')
ax.legend(loc='best', fontsize=16)
ax.set_xlim(100, 1000)
ax.set_ylim(0, 72000)
ax.set_title('Espectros de masas', fontsize=20)
ax.set_xlabel('m/z', fontsize=16)
ax.set_ylabel('Intensidad', fontsize=16)
ax.tick_params(labelsize=13, labelcolor='k', width=1)
plt.grid()
```
