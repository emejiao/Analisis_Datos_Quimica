```python
A1=(ys[450:550]-np.min(ys[0:,0])).sum(axis=0)
C1=pd.DataFrame([1145, 11450, 18320, 25190, 35000, 45000, 6870])
A1=pd.DataFrame(A1)
D1=pd.concat([C1, A1], axis=1)
D1=D1.set_axis(['[ppm]', 'Int'], axis=1, inplace=False)

A2=(ys1[340:380]-np.min(ys1[0:,0])).sum(axis=0)
C2=pd.DataFrame([1122.2, 14427.5, 18836, 4008, 9217])
A2=pd.DataFrame(A2)
D2=pd.concat([C2, A2], axis=1)
D2=D2.set_axis(['[ppm]', 'Int'], axis=1, inplace=False)

A3=(ys2[630:710]-np.min(ys2[0:,0])).sum(axis=0)
C3=pd.DataFrame([136844, 3909.8, 7037.6, 977.4, 9779.5])
A3=pd.DataFrame(A3)
D3=pd.concat([C3, A3], axis=1)
D3=D3.set_axis(['[ppm]', 'Int'], axis=1, inplace=False)

A4=(ys3[480:545]-np.min(ys3[0:,0])).sum(axis=0)
C4=pd.DataFrame([1142, 14097, 2430, 5833, 9772])
A4=pd.DataFrame(A4)
D4=pd.concat([C4, A4], axis=1)
D4=D4.set_axis(['[ppm]', 'Int'], axis=1, inplace=False)
#D1, D2, D3, D4
```

<font color=k size=6> **4. Regresi√≥n lineal y coeficiente de determinaci√≥n $R^2$**</font>  

---

```python
import sklearn.metrics as skm
from sklearn import linear_model
x=D1[['[ppm]']] 
y=D1[['Int']]
reg = linear_model.LinearRegression()
reg.fit(x, y)
y1_pred = reg.predict(x)
m=reg.coef_[0]
b=reg.intercept_[0]
mse=skm.mean_squared_error(y, y1_pred)
r2=skm.r2_score(y, y1_pred)

x1=D2[['[ppm]']] 
y1=D2[['Int']]
reg1 = linear_model.LinearRegression()
reg1.fit(x1, y1)
y2_pred = reg1.predict(x1)
m1=reg1.coef_[0]
b1=reg1.intercept_[0]
mse1=skm.mean_squared_error(y1, y2_pred)
r22=skm.r2_score(y1, y2_pred)

x2=D3[['[ppm]']] 
y2=D3[['Int']]
reg2 = linear_model.LinearRegression()
reg2.fit(x2, y2)
y3_pred = reg2.predict(x2)
m2=reg2.coef_[0]
b2=reg2.intercept_[0]
mse2=skm.mean_squared_error(y2, y3_pred)
r222=skm.r2_score(y2, y3_pred)

x3=D4[['[ppm]']] 
y3=D4[['Int']]
reg3 = linear_model.LinearRegression()
reg3.fit(x3, y3)
y4_pred = reg3.predict(x3)
m3=reg3.coef_[0]
b3=reg3.intercept_[0]
mse3=skm.mean_squared_error(y3, y4_pred)
r2222=skm.r2_score(y3, y4_pred)
```
