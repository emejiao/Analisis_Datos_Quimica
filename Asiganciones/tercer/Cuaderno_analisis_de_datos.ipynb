```python
Lugar2 = ("D:\\Cristian\\Usb\\USB_Universidad\\Analisis_de_Datos\\Datos\\IR2.csv")
Lugar3 = ("D:\\Cristian\\Usb\\USB_Universidad\\Analisis_de_Datos\\Datos\\IR3.csv")
Lugar4 = ("D:\\Cristian\\Usb\\USB_Universidad\\Analisis_de_Datos\\Datos\\IR4.csv")
Lugar5 = ("D:\\Cristian\\Usb\\USB_Universidad\\Analisis_de_Datos\\Datos\\IR5.csv")
Lugar6 = ("D:\\Cristian\\Usb\\USB_Universidad\\Analisis_de_Datos\\Datos\\IR6.csv")
IR2 = np.loadtxt(Lugar2, delimiter=',')
IR3 = np.loadtxt(Lugar3, delimiter=',')
IR4 = np.loadtxt(Lugar4, delimiter=',')
IR5 = np.loadtxt(Lugar5, delimiter=',')
IR6 = np.loadtxt(Lugar6, delimiter=',')
IR6.shape
```

```python
Ab1=2-np.log10(IR1[1:, 1])
Ab2=2-np.log10(IR2[1:, 1])
Ab3=2-np.log10(IR3[1:, 1])
Ab4=2-np.log10(IR4[1:, 1])
Ab5=2-np.log10(IR5[1:, 1])
Ab6=2-np.log10(IR6[1:, 1])
```

```python
fig, ax= plt.subplots(figsize=(12,8))

plt.rcParams["font.family"] = 'Times New Roman'
ax.plot(IR1[1:, 0], Ab1, 'b')
ax.plot(IR2[1:, 0], Ab2, 'k')
ax.plot(IR3[1:, 0], Ab3, 'r')
ax.plot(IR4[1:, 0], Ab4, 'g')
ax.plot(IR5[1:, 0], Ab5, 'c')
ax.plot(IR6[1:, 0], Ab6, 'm')
ax.set_xlim(400, 4000)
ax.set_ylim(0.00, 0.50)
ax.set_title("IRs")
ax.set_xlabel("Número de onda [cm^-1]")
ax.set_ylabel("Absorbancia")
fig.savefig("IRs.jpg", dpi = 140)
```
