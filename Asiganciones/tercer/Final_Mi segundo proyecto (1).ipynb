```python
datafile="./Datos/Isoc-Chrom.xlsx"#este archivo contiene tres hojas de calculo 
data = pd.read_excel(datafile)
data
```

```python
# Librería numpy permite obtener la pendiente y el corte de un ajuste lineal entre dos serie datos
import numpy as np
x=data['Mol-Isoctano']
y=data['Area-Pico']
# Ajuste a una recta (polinomio de grado 1)
p = np.polyfit(x, y, 1) 
#lo hacemos con numpy y este usa la acció "polifit" que me permite hacer ajustes y en este caso de orden 1 (es decir ajuste lineal)
# No es posible obtener métricas estadísticas con el coeficiente de regresión
print(p)
# y(x)=p0x+p1=1.00000009e+00x-4.29084910e-05, muestra la pendiente y el corte
```

```python
fig, ax= plt.subplots()
c=np.arange(0,2,0.01)
plt.scatter(x, y)
plt.plot(c, 2.09*c+0.26)
ax.set_title('Curva de calibración')
ax.set_xlabel("Mol-Isoctano")
ax.set_ylabel("Area-Pico")
```

```python
fig, ax= plt.subplots()
ax.scatter(x, y, label='Original data', edgecolors='black')
ax.plot(x, y, 'r', lw=2, label='Curva calibración')
ax.legend()
ax.set_title("Model_1")
ax.set_xlabel("Predicted (%)")
ax.set_ylabel("Measured (%)")
#ax.text(12,7, r"$std-err = "+str("{0:.3f}".format(std_err))+r"$", fontsize=10, color="black")
#ax.text(12,6.0, r"$r^2 = "+str("{0:.3f}".format(r_value**2))+r"$", fontsize=10, color="black")
plt.show()
#fig.savefig("ccal.png",bbox_inches='tight')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sio
```

```python
import glob
CurvaCa = pd.DataFrame()
#nombres=['MF']
for f in glob.glob("./Datos/Ca/curva1/*c1.asc"):
    df = pd.read_csv(f, header=None, delimiter='\t' )
    CurvaCa = pd.concat([CurvaCa, df], axis=1)
    print(f)
data1=CurvaCa.drop(CurvaCa.index[[0]]) # Eliminamos el primer la primera fila
ys1=data1[1]
data1.columns=range(data1.shape[1])
xs1=data1[[0]] #solo una columna de long de onda para todas
ys1.columns=range(ys1.shape[1])
```

```python
xs1
```

```python
ys1=np.array(ys1)
xs1=np.array(xs1)
fig, esp= plt.subplots(figsize=(12, 6))
plt.rcParams["font.family"] = 'Times New Roman'
esp.plot(xs1, ys1[:,0])
esp.set_xlim(np.min(xs1)+10, np.max(xs1)-10)
esp.set_title("Spectra Ca")
esp.set_xlabel("Wavelenght (nm)")
esp.set_ylabel("Intensity (arb. units)");
```

```python
plt.plot(ys1[320:410]-np.min(ys1[0:,0]));
```

```python
A1=(ys1[320:410]-np.min(ys1[0:,0])).sum(axis=0)
C1=pd.DataFrame([1122.2, 14427.5, 18836, 4008, 9217])
A1=pd.DataFrame(A1)
D1=pd.concat([C1, A1], axis=1)
D1=D1.set_axis(['[ppm]', 'Int'], axis=1, inplace=False)
D1
```

```python
import sklearn.metrics as skm
from sklearn import linear_model
x1=D1[['[ppm]']] 
y1=D1[['Int']]
reg = linear_model.LinearRegression()
reg.fit(x1, y1)
y1_pred = reg.predict(x1)
m=reg.coef_[0]
b=reg.intercept_[0]
mse=skm.mean_squared_error(y1, y1_pred)
r2=skm.r2_score(y1, y1_pred)
```

```python
fig, ax= plt.subplots()
ax.scatter(x1, y1, color='green', label='Datos originales', edgecolors='black')
ax.plot(x1, y1_pred, 'r', label='Ajuste')
ax.legend()
ax.set_title("curva de Calibración Ca")
ax.set_xlabel("Concentración (ppm)")
ax.set_ylabel("Intensidad (units arb.)")
ax.text(10000,0.4e6, r"$r^2 = "+str("{0:.3f}".format(r2))+r"$", fontsize=12, color="black")
plt.show()
fig.savefig("CurvaCa.pdf",bbox_inches='tight')
```
