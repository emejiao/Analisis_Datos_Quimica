```python
data1= pd.read_csv("https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR1.CSV")
data1.head(10)
 
```

```python
Data1_sorted= data1.sort_values(by = '3.996813e+002')
Data1_sorted.head()
```

```python
Data1_sorted.sort_values(by = '3.996813e+002',ascending = False , inplace= True)
Data1_sorted.head()
```

```python
data1['3.996813e+002'].max()
```

```python
data1['3.996813e+002'].min()
```

```python
COL_1= data1['3.996813e+002']

COL_1.head()
```

```python
COL_2= data1['0.000000e+000']
COL_2.head()
```

```python
plt.plot(Data1_sorted, COL_2, 'ro-', lw=2, )
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='best')
plt.grid()
plt.savefig('fig1.jpg', dpi=250)
```

```python
fig, ax = plt.subplots()

ax.plot(COL_1, COL_2, '#551A8B')
ax.set_xlim(4000,500)
ax.set_ylim(30, 102)
ax.set_title("Espectro IR")
ax.set_xlabel("x")
ax.set_ylabel("y")
```
