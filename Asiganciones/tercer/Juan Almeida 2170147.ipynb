```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lugarIR1=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR1.CSV')
IR1 = pd.read_csv(lugarIR1, header= None)
```

```python
lugarIR2=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR2.CSV')
IR2 = pd.read_csv(lugarIR2, header= None)
```

```python
lugarIR3=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR3.CSV')
IR3 = pd.read_csv(lugarIR3, header= None)
```

```python
lugarMS1=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS.txt')
MS1np = np.loadtxt(lugarMS1)
```

```python
lugarMS3=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS3.txt')
MS3np = np.loadtxt(lugarMS3)
```

```python
MS1 = pd.DataFrame(MS1np)
MS2 = pd.DataFrame(MS2np)
MS3 = pd.DataFrame(MS3np)
```

```python
fig, ax = plt.subplots()

ax.plot(IR1[0], IR1[1], 'y')
ax.set_title("Espectro de IR")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Transmitancia")
```

```python
fig, ax = plt.subplots()

ax.plot(IR2[0], IR2[1], 'y')
ax.set_title("Espectro de IR")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Transmitancia")
```

```python
fig, ax = plt.subplots()

ax.plot(IR3[0], IR3[1], 'y')
ax.set_title("Espectro de IR")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Transmitancia")
```

```python
fig, ax = plt.subplots()

ax.plot(IR1[0], IR1[1], 'y')
ax.plot(IR2[0], IR2[1], 'g')
ax.plot(IR3[0], IR3[1], 'k')
ax.set_title("Espectro de IR")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Transmitancia")
```
