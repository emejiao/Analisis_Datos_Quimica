```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pathIR1=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR1.CSV')
pathIR2=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR2.CSV')
pathIR3=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/IR3.CSV')
IR1 = pd.read_csv(pathIR1, header= None)
IR2 = pd.read_csv(pathIR2, header= None)
IR3 = pd.read_csv(pathIR3, header= None)
pathMS1=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS.txt')
pathMS2=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS2.txt')
pathMS3=('https://github.com/emejiao/Analisis_Datos_Quimica/raw/main/Datos/MS3.txt')
MS1np = np.loadtxt(pathMS1)
MS2np = np.loadtxt(pathMS2)
MS3np = np.loadtxt(pathMS3)
MS1 = pd.DataFrame(MS1np)
MS2 = pd.DataFrame(MS2np)
MS3 = pd.DataFrame(MS3np)
```

```python
fig, ax = plt.subplots()

ax.plot(MS1[0], MS1[1], 'g')
ax.plot(MS2[0], MS2[1], 'b')
ax.plot(MS3[0], MS3[1], 'r')
ax.set_title("Espectro de masas")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Intensidad (u. arb.)")
```

```python
fig, ax = plt.subplots()

ax.plot(IR1[0], IR1[1], 'g')
ax.plot(IR2[0], IR2[1], 'r')
ax.plot(IR3[0], IR3[1], 'b')
ax.set_title("Espectro de IR")
ax.set_xlabel("Longitud de onda (nm)")
ax.set_ylabel("Transmitancia (u. arb.)")
```
