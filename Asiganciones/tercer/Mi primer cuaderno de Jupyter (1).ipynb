<font color='#8800C4' size=5>         **Gráficas IR**

```python
#Una sola gráfica
fig, ax= plt.subplots(figsize=(20,6))
plt.plot(IR1['m/z'],IR1['Int'], color = 'red', linestyle='-' )
ax.set_xlim(500, 4000)
ax.set_title('IR1', fontsize = 20)
plt.xlabel('Número de onda ($cm^{-1}$)',  fontsize=15)
plt.ylabel('Transmitancia', fontsize=15)
```

```python
#3 gráficas
fig, ax= plt.subplots(figsize=(20,6))
plt.plot(IR1['m/z'],IR1['Int'], color = 'red', linestyle='-', label = 'IR1' )
plt.plot(IR2['m/z'],IR2['Int'], color = 'green', linestyle='-', label = 'IR2' )
plt.plot(IR3['m/z'],IR3['Int'], color = 'blue', linestyle='-', label = 'IR3' )
ax.legend(loc='best', fontsize=15)
ax.set_xlim(500, 4000)
ax.set_title('IR1, IR2, IR3', fontsize = 20)
plt.xlabel('Número de onda ($cm^{-1}$)',  fontsize=15)
plt.ylabel('Transmitancia', fontsize=15)
fig.savefig('Mi primera gráfica.png', dpi=200)
```
