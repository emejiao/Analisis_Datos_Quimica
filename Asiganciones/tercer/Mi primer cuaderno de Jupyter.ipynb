```python
#Combinación archivos IR
IRc = pd.concat([IR1, IR2[['Int']], IR3[['Int']]], axis = 1)
IRc.head()
```

```python
#Cambio de nombre de las columnas
IRc.set_axis(['Wn', 'T1', 'T2', 'T3'], axis='columns', inplace=True)
IRc.head()
```

```python
#Combinación archivos MS
MSc = pd.concat([MS1, MS2[['Y']], MS3[['Y']]], axis = 1)
MSc.head()
```

```python
#Cambio de nombre de las columnas
MSc.set_axis(['m/z', 'I1', 'I2', 'I3'], axis = 'columns', inplace = True)
MSc.head()
```

```python
#Creación archivo excel con los datos de IR y MS
Excel = pd.ExcelWriter('IRMS.xlsx', engine='xlsxwriter')
IRc.to_excel(Excel, sheet_name = 'IR')
MSc.to_excel(Excel, sheet_name = 'MS')
Excel.save()
```
